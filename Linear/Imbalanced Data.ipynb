# K thut over-sampling v under-sampling

Vic x l d liu mt cn bng (Imbalanced Data) l rt quan trng  m hnh hc my khng b thin v (bias) v lp a s. K thut ly mu li (Resampling) l phng php ph bin nht  gii quyt vn  ny, bao gm hai nhm chnh l Oversampling (Tng mu) v Undersampling (Gim mu).  
## 1. Oversampling (K thut Tng mu)   
K thut Oversampling tp trung vo vic tng s lng mu ca lp thiu s (Minority Class)  cn bng vi lp a s (Majority Class).  
### A. Random Oversampling (Tng mu Ngu nhin)
Cch thc: Sao chp ngu nhin cc mu hin c trong lp thiu s v thm chng vo tp d liu hun luyn.  
u im: n gin, d thc hin.  
Nhc im: D dn n Overfitting (hc thuc lng) v m hnh c hun luyn lp i lp li trn cng mt tp d liu nh.
### B. SMOTE (Synthetic Minority Oversampling Technique) 
Cch thc: y l phng php ph bin nht. Thay v sao chp, SMOTE tng hp (to ra) cc mu mi cho lp thiu s.  
1. Chn mt mu A thuc lp thiu s.
2. Tm k hng xm gn nht ca A (Nearest Neighbors).  
3. To mu mi bng cch ni suy (interpolation) gia mu A v mt trong cc hng xm ca n (v d: ly trung im trn on thng ni hai mu).

u im: Gim nguy c Overfitting so vi Random Oversampling v cc mu mi c to ra hi khc so vi mu gc.  
Nhc im: C th to ra cc mu "ngoi lai" (noise) nu lp thiu s qu tha tht.
## 2. Undersampling (K thut Gim mu) 
K thut Undersampling tp trung vo vic gim s lng mu ca lp a s (Majority Class).  
### A. Random Undersampling (Gim mu Ngu nhin)
Cch thc: Xa b ngu nhin cc mu t lp a s cho n khi lp a s cn bng vi lp thiu s.  
u im: n gin, gip gim thi gian hun luyn m hnh.  
Nhc im: C th dn n mt mt thng tin quan trng nu cc mu b loi b cha d liu gi tr.
### B. Tomek Links
Cch thc: y l mt k thut lm sch d liu da trn khong cch.Xc nh cc cp mu Tomek Links: mt mu thuc lp thiu s v mt mu thuc lp a s l hng xm gn nht ca nhau, v chng c nhn khc nhau.Loi b mu thuc lp a s trong cp Tomek Link .  
u im: Xa b cc mu a s nm gn ranh gii ca lp thiu s, gip ranh gii phn loi tr nn r rng hn.
