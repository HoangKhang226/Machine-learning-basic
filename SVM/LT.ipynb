SVM là gì?
Support Vector Machine (SVM) là một thuật toán tìm kiếm siêu mặt phẳng (hyperplane) tốt nhất để phân tách dữ liệu thành các lớp khác nhau. Mục tiêu chính của SVM là tối đa hóa khoảng cách (margin) giữa siêu mặt phẳng và các điểm dữ liệu gần nhất của mỗi lớp.

Ba Khái niệm Cốt lõi
1. Siêu Mặt phẳng (Hyperplane)  
Đây là đường hoặc mặt phẳng phân chia các lớp dữ liệu.

Trong không gian 2D, siêu mặt phẳng là một đường thẳng.

Trong không gian 3D, siêu mặt phẳng là một mặt phẳng.

Trong không gian nhiều chiều hơn, nó được gọi là siêu mặt phẳng.  

2. Khoảng cách Tối đa (Maximum Margin)  
SVM tìm siêu mặt phẳng có khoảng cách lớn nhất đến các điểm dữ liệu gần nhất của bất kỳ lớp nào. Khoảng cách này được gọi là margin.

Siêu mặt phẳng với margin càng lớn thì khả năng tổng quát hóa (generalization) của mô hình càng tốt, giúp nó hoạt động hiệu quả hơn với dữ liệu mới chưa từng thấy.

3. Vector Hỗ trợ (Support Vectors)  
Đây là các điểm dữ liệu nằm gần nhất với siêu mặt phẳng (nằm trên margin).

Chỉ có các điểm này mới cần thiết để định nghĩa vị trí và hướng của siêu mặt phẳng. Nếu bạn loại bỏ bất kỳ điểm dữ liệu nào khác (không phải Support Vector), siêu mặt phẳng vẫn giữ nguyên. Điều này làm cho SVM rất hiệu quả về mặt tính toán.
