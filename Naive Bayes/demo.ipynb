```python
import pandas as pd
import numpy as np 
```

```python
X = np.array([["Red", "Sports", "Domestic"],
              ["Red", "Sports", "Domestic"],
              ["Red", "Sports", "Domestic"],
              ["Yellow", "Sports", "Domestic"],
              ["Yellow", "Sports", "Imported"],
              ["Yellow", "SUV", "Imported"],
              ["Yellow", "SUV", "Imported"],
              ["Yellow", "SUV", "Domestic"],
              ["Red", "SUV", "Imported"],
              ["Red", "Sports", "Imported"],
              ["Blue", "Family", "Domestic"]])

Y = np.array(["Yes", "No", "Yes", "No", "Yes", "No", "Yes", "No", "No", "Yes","Yes"])
```

```python
X=pd.DataFrame(X,columns=["color","type","origin"])
Y=pd.DataFrame(Y,columns=["stolen"])
X.head()
```

```python
X=pd.get_dummies(X)
X.head()
```

```python
x_new=[[0,1,0,1,1,1,0,0]]
```

# Bernoulli

```python
from sklearn.naive_bayes import BernoulliNB
```

```python
model_B=BernoulliNB().fit(X,Y)
```

```python
y_pred=model_B.predict(x_new)
model_B.predict_proba(x_new), y_pred
```

# Gauss

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model_G=GaussianNB().fit(X,Y)
```

```python
y_pred=model_G.predict(x_new)
model_G.predict_proba(x_new), y_pred
```
