# SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python
X_s,Y_s=SMOTE().fit_resample(df[["0","1"]],df.target)
Y_s.value_counts()
```

```python
df_S=pd.DataFrame(X_s)
df_S["target"]=Y_s
df_S.head()
```

```python
df_S.target.value_counts()
```
